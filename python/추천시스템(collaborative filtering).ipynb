#### 전처리

```python
rating_data.drop('timestamp',axis = 1, inplace = True)
```

```python
user_movie_rating = pd.merge(rating_data, movie_data, on = 'movieId')
```

```python
#pivot table 만들기(아이템 기반 협업 필터링 수행하기 위해)
#사용자-영화에 따른 평점 점수가 데이터로 들어가야함
```
