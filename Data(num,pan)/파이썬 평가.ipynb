## 3번

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#1) qcut 명령으로 세 개의 나이 그룹을 만든다.
```

```python
titanic['age_group']=pd.qcut(titanic['age'],3,labels=['노년','중년','소년']).astype(str)
```

```python
#2) 성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다.
#행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.
#생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다.
titanic.head()
```

```python
#방법1
#생존률 계산 함수
def survived_rate(group):
    return group.survived[group.survived==1].count() / group.survived.count()
    
    
```

```python
titanic.groupby(['sex','age_group','class']).apply(survived_rate).unstack()


```

```python
#방법2
#생존률 계산 함수
def survived_rate(s):
    count=0
    for i in s:
        if i == 1:
            count += 1
    return count / s.count()
```

```python
#사용자 정의함수 호출(agg)하여 적용
titanic.groupby(['sex','age_group','class'])[['survived']].agg(survived_rate).unstack()



```

```python
#3)성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.
```

```python
titanic.pivot_table('survived','sex','class')
```

```python
from sklearn.datasets import load_digits
X = load_digits().data
```

```python
#1)내적을 이용하여 첫 번째 이미지와 10번째 이미지의 유사도를 구하라.
image1=X[0]
image9=X[9]
```

```python
np.dot(image1,image9)
```

```python
#2)코사인유사도를 이용하여 첫 번째 이미지와 10번째 이미지의 유사도를 구하라.
import numpy as np
def cos_similarity(x, y, eps=1e-8):
    nx = x / (np.sqrt(np.sum(x ** 2)) + eps)
    ny = y / (np.sqrt(np.sum(y ** 2)) + eps)
    return np.dot(nx, ny)

```

```python
cos_similarity(image1,image9)
```

```python
df=pd.read_csv('0402_test.csv',parse_dates=['datetime'])
df
```

```python
df['year']=df['datetime'].dt.year
df['month']=df['datetime'].dt.month
df['day']=df['datetime'].dt.day
df['hour']=df['datetime'].dt.hour
```

```python
df
```

```python
#1) 각 컬럼값에 따른 count값의 변화를 시각화 하시오.
```
