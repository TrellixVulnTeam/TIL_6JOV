```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
years=range(1880,2011)
pieces=[]

for year in years:
    path="babynames/yob%d.txt" % year
    frame=pd.read_csv(path, header=None, names=['name', 'sex', 'births'])
    frame['year']=year
    pieces.append(frame)
    #print(frame)
    #print(path)
    
len(pieces[0][pieces[0].sex=='F']) #942
len(pieces[130][pieces[130].sex=='F']) #19698
```

```python
len(pieces)
```

```python
names=pd.concat(pieces, ignore_index=True)
names
```

```python
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
#1.
#연도 및 성별로 그룹화 -> 태어난 아이의 수 -> 시각화
#선그래프(남/여)
#x축 year, y축  태어난 아이의 수 합계
```

```python
year_sex= names.pivot_table('births',index='year',columns='sex',aggfunc=sum)
year_sex.plot()
plt.title('연도별 태어난 아이 수')


```

```python
#2.
# names에 prop열 추가
# prop열은 연도별 각각의 이름에 해당되는 아이의 수 /  연도에 태어난 전체 아이의 수
# ex) 1880년 전체 아이의 수 100명, john이름 아이의 수 5명
# prop=5/100= 0.05
```

```python
year_sex_group=names.groupby(['year','sex'])
year_sex_group

def add_prop(group):
    births = group.births
    
    group['prop'] = births / births.sum()
    return group

names= year_sex_group.apply(add_prop)
names.head()
```

```python
year_sex_group=names.groupby(['year'])
year_sex_group

def add_prop(group):
    births = group.births
    
    group['prop'] = births / births.sum()
    return group

names= year_sex_group.apply(add_prop)
names.head()
```

```python
#3.
# names에서 연도별, 성별 가장 많은 이름을 1000개씩 추출
# 1880년 여자 메리, 애나,....마린
# 1880년 남자 존, 마이클,....
```

```python

top1000=year_sex_group.apply(lambda x:x.sort_values(by='births' , ascending=False)[:1000])
top1000.head(10)
```

```python
#4
#John, Mary, Harry 이름의 유행 변화 ???
# John, Mary, Harry에 대해 시각화(선 그래프)
# x축 year, y축 태어난아이의수

#이름의 다양성 변화?  분석(여러분의 아이디어)
#1880년 여자 아이들의 이름의 종류:942,
#2010년 여자 아이들의 이름의 종류:19698
#이름이 매우 다양해졌다
#시각화, x축 year, y축 이름의 종류 수, 선그래프, 성별
```

```python
john = names[names['name']=='John'].groupby('year').sum()
john
```

```python
mary = names[names['name']=='Mary'].groupby('year').sum()
mary
```

```python
harry = names[names['name']=='Harry'].groupby('year').sum()
harry
```

```python
john.index
```

```python
plt.plot(john.index, john, 'r-', 
         mary.index, mary, 'b-',
         harry.index, harry, 'g-')
plt.title("이름의 유행 변화")
plt.ylabel("아이의 수(명)")
plt.legend(("john","mary","harry"), loc=2)
plt.show()
```

```python
year_n_f = names[names['sex']=='F'].groupby('year').size()
year_n_m = names[names['sex']=='M'].groupby('year').size()
```

```python
pf = year_n_f.plot()
pm = year_n_m.plot()
plt.title("연도별 아이의 이름 종류")
plt.ylabel("이름 종류(개)")
plt.legend(("female","male"), loc=2)
plt.show()
```

```python
diff = abs(year_n_f.values - year_n_m.values)
print(max(diff))  #6157: 여자아이와 남자아이 이름 종류의 최대격차
print(year_n_f.index[np.where(diff==max(diff))][0])  #2007: 가장 차이 큰 해
```

```python
#이름의 첫번째 글자

names['first_letters']=names['name'].apply(lambda x: x[0])
#성별 선호도
first_name_by_gender=names.pivot_table(index='first_letters',
                                       columns='sex',values='births',aggfunc=sum)
first_name_by_gender
```

```python
fig, axes=plt.subplots(figsize=(20,10))
first_name_by_gender.plot(kind='bar',rot=0, ax=axes)
```
